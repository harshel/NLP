```python
from bs4 import BeautifulSoup
import urllib.request
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
import string
from collections import Counter
import operator
from tqdm import tqdm_notebook
```

```python
def parse_article(link):
    link=link
    req=urllib.request.Request(link)
    response=urllib.request.urlopen(req)
    the_page=response.read()
    soup=BeautifulSoup(the_page,'lxml')
    text=' '.join([p.text for p in soup.findAll('p')])
    sent_tokens = sent_tokenize(text)
    corpus1 = []
    for sent in tqdm_notebook(sent_tokens):
        word_tokens = word_tokenize(sent)
        data = [w.lower() for w in word_tokens]
        data = [w for w in data if w not in stopwords.words('english')]
        data = [w for w in data if w not in string.punctuation]
        data = [w for w in data if w.isalpha()]
        data = ' '.join([w for w in data])
        corpus1.append(data)
    corpus2=word_tokenize(' '.join(corpus1))
    x=dict(Counter(corpus2))
    sorted_x = sorted(x.items(), key=operator.itemgetter(1),reverse=True)
    

    return sorted_x
```

```python
article1=dict(parse_article('https://www.thehindu.com/business/Industry/emission-scandal-ngt-slaps-500-crore-fine-on-volkswagen/article26454437.ece'))
article2=dict(parse_article('https://www.nytimes.com/2019/03/15/business/volkswagen-winterkorn-sec-fraud.html'))
article3=dict(parse_article('https://en.wikipedia.org/wiki/Volkswagen_emissions_scandal'))
```
