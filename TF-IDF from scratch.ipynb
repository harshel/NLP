```python
import pandas as pd
```

```python
df=pd.read_json('winemag-data-130k-v2.json')
df.head()
```

```python
df=df[['description']]
df.head()
```

```python
df = df.dropna()
```

```python
df["description"] = df["description"].str.lower()
df.head()
```

```python
from nltk.tokenize import word_tokenize as wtk
from nltk.stem import WordNetLemmatizer as wnl
from nltk.corpus import stopwords
from string import punctuation
lem = wnl()
custom = list(set(stopwords.words('english')))+list(punctuation)+['``', "'s", "...", "n't"]
def lemon(narr):
    words = wtk(narr)
    words = [lem.lemmatize(x) for x in words]
    bow = [x for x in words if x not in custom]
    narr = ' '.join([x for x in words])
    return narr
```

```python
df2['description']=df2['description'].apply(lemon)
```
